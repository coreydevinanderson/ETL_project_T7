```python
job_data = job_data_csv.drop(columns=['crawl_timestamp','url',
                                      'job_description','geo','cursor','contact_email',
                                      'contact_phone_number','uniq_id','html_job_description',"inferred_state",
                                      "inferred_city",'inferred_country'])

job_data.head()
```

```python
#drop all rows with null values in city, state
job_data.dropna(subset = ["city","state"], inplace=True)
job_data
```

```python
#clean state names: that conatins'work-from-home, computer,united' ---> remote
loc = ['work','computer','united']

for x in loc:
    
    wfh = job_data[job_data['state'].str.contains(x, case=False)]
    wfh_list = wfh['state'].to_list()
    job_data["state"].replace(wfh_list,"Remote",inplace=True)
```

```python
nj = job_data[job_data['state'].str.contains("jersey", case=False)]
nj_list = nj['state'].to_list()
job_data["state"].replace(nj_list,"NJ",inplace=True)
```

```python
### state names
abbrev = ['New','virginia','Hawaii','Idaho','Louisiana','Washington','Massachusetts']
replace_abbrev = ["NM","VA","HI","ID","LA","WA","MA"]

for x in range(7):
    
    abb = job_data[job_data['state'].str.contains(abbrev[x], case=False)]
    abb_list = abb['state'].to_list()
    job_data["state"].replace(abb_list,replace_abbrev[x],inplace=True)
```
