```python
 clean_job_data = pd.read_csv("output_csv/clean_job_data.csv")
```

```python
display(clean_job_data)
```

```python
cleaner_job_data = clean_job_data.copy()
```

```python
cleaner_job_data = cleaner_job_data.drop(["country"], axis = 1)
```

```python
even_cleaner_job_data = cleaner_job_data.loc[cleaner_job_data["state"] != "PR"]
```

```python
display(even_cleaner_job_data)
```

```python
cleaner_job_data_NULL = even_cleaner_job_data.fillna('NULL')
```

```python
cleaner_no_remote = cleaner_job_data_NULL.loc[cleaner_job_data_NULL["state"] != "Remote"]
```

```python
cleaner_no_remote = cleaner_no_remote.loc[cleaner_no_remote["state"] != "NN"]
```

```python
cleaner_no_remote.reset_index(drop = True, inplace = True)
```

```python
display(cleaner_no_remote)
```

```python
cleaner_no_remote.to_csv("output_csv/cleaner_job_data.csv", index = False)

# Read for postgreSQL
```
