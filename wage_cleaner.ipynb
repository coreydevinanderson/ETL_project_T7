```python
import pandas as pd
```

```python
salary = pd.read_csv("output_csv/clean_salary_data.csv")
```

```python
display(salary)
```

```python
salary.drop(["index", "State"], axis= 1, inplace = True)
```

```python
salary.reset_index(drop = True, inplace = True)
```

```python
display(salary)
```

```python
salary_df = salary.rename(columns = {"State Abbrev":"state", 
                                     "Job Title": "job_title",
                                     "Hourly Wage Mean": "hourly_wage_mean",
                                     "Hourly Wage Median": "hourly_wage_median",
                                     "Annual Wage Median": "annual_wage_median"})
```

```python
salary_df.fillna("NULL", inplace = True)
```

```python
display(salary_df)
```

```python
nums_raw = salary_df["annual_wage_median"].to_list()
```

```python
salary_df["annual_wage_median"] = [int(number.replace(",", "")) for number in nums_raw]
```

```python
display(salary_df)
```

```python
salary_df.to_csv("output_csv/salary_cleaner.csv", index = False)
```
