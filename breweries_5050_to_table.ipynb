```python
# Dependencies
import requests
import json
import pandas as pd
```

<strong>Use state names as search string in call to openbrewerydb.org</strong>

<p>Must convert to lower case...then add underscore in place of space.</p>

```python
state_names = ["Alaska", "Alabama", "Arkansas", "American Samoa", "Arizona", "California", "Colorado", "Connecticut", "District of Columbia", "Delaware", "Florida", "Georgia", "Guam", "Hawaii", "Iowa", "Idaho", "Illinois", "Indiana", "Kansas", "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", "Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", "North Dakota", "Nebraska", "New Hampshire", "New Jersey", "New Mexico", "Nevada", "New York", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Puerto Rico", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Virginia", "Virgin Islands", "Vermont", "Washington", "Wisconsin", "West Virginia", "Wyoming"]

state_lower = [state.lower() for state in state_names] 
```

```python
state_lower_under = [state.replace(" ", "_") for state in state_lower]
```

```python
print(state_lower_under)
```

```python
len(state_lower_under)
```

```python
# Do not include

territories = ["american_samoa", "guam", "puerto_rico", "virgin_islands"]

slu_51 = [state for state in state_lower_under if state not in territories]
```

```python
len(slu_51)
```

```python
print(slu_51)
```

```python
target_url = "https://api.openbrewerydb.org/breweries?by_state="
```

```python
# Example url with state 14 in alphabetical list of states

state_url = f"{target_url}{slu_51[13]}&per_page=50"
print(state_url)
```

<h3> Example call to API </h3>

```python
# Run a request to endpoint and convert result to json
geo_data = requests.get(state_url).json()

# Print the json
print(geo_data)

print(json.dumps(geo_data, indent = 4, sort_keys = True))

# Looks like a list of dictionaries
```

```python
local_count = 0


# Make the DataFrame

brew_df = pd.DataFrame({"id":[],
                        "obdb_id":[],
                        "name":[],
                        "brewery_type":[],
                        "city":[],
                        "state":[],
                        "country":[],
                        "postal_code":[],
                        "longitude":[],
                        "latitude":[], 
                        "website_url":[]})


for i in range(len(slu_51)):
    
    target_url = "https://api.openbrewerydb.org/breweries?by_state="
    state_url = f"{target_url}{slu_51[i]}&per_page=50"
    geo_data = requests.get(state_url).json()
    
    for j in range(len(geo_data)):
        brew_df.loc[local_count, "id"] = geo_data[j]["id"]
        brew_df.loc[local_count, "obdb_id"] = geo_data[j]["obdb_id"]
        brew_df.loc[local_count, "name"] = geo_data[j]["name"]
        brew_df.loc[local_count, "brewery_type"] = geo_data[j]["brewery_type"]
        brew_df.loc[local_count, "city"] = geo_data[j]['city']
        brew_df.loc[local_count, "state"] = geo_data[j]['state']
        brew_df.loc[local_count, "country"] = geo_data[j]['country']
        brew_df.loc[local_count, "postal_code"] = geo_data[j]['postal_code']
        brew_df.loc[local_count, "longitude"] = geo_data[j]['longitude']
        brew_df.loc[local_count, "latitude"] = geo_data[j]['latitude']
        brew_df.loc[local_count, "website_url"] = geo_data[j]['website_url']
        
        local_count += 1
    
```

```python
brew_df_culled = brew_df.drop(["id", "obdb_id", "country"], axis = 1)
```

```python
brew_df_culled.reset_index(drop = True, inplace = True)
brew_df_culled.fillna('NULL', inplace = True)
```
