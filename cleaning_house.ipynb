```python
import pandas as pd
import os 
import csv

filepath = "data/dirty_zillow_homevalues_SFR_state.csv"
```

```python
complete_housing_data = pd.read_csv(filepath, delimiter=",")

complete_housing_data
```

```python
columns_2019 = []

for nineteen in complete_housing_data.columns:
    if nineteen.endswith('/19'):
        columns_2019.append(nineteen)
columns_2019
```

```python
housing_data_2019 = complete_housing_data[['RegionName','StateName', '1/31/19',
 '2/28/19',
 '3/31/19',
 '4/30/19',
 '5/31/19',
 '6/30/19',
 '7/31/19',
 '8/31/19',
 '9/30/19',
 '10/31/19',
 '11/30/19',
 '12/31/19']]
housing_data_2019
```

```python
avg_value_2019 = housing_data_2019.mean(axis=1)
avg_value_2019
```

```python
add_avg = housing_data_2019.assign(AverageHomePrice = avg_value_2019)

add_avg
```

```python
df_state_price_2019 = add_avg[['StateName', 'AverageHomePrice']]
df_state_price_2019
```

```python
pd.options.display.float_format = '${:,.0f}'.format
```

```python
df_state_price_2019
```

```python
df_state_price_2019.rename(columns={'AverageHomePrice': 'Average Home Price', 'StateName': 'State'}, inplace=True)
```
